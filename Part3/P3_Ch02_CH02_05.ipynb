## 라이브러리 import

## figure와 subplot


matplotlib에서 그래프는 figure라는 큰 틀에 ax라는 subplot을 추가해서 만듭니다. 

```python
# figure instancee 생성
fig = plt.figure()
plt.show()
```

```python
# 빈 subplot생성
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# figure size 조절
fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot()
plt.show()
```

```python
# 여러개의 subplot 만들기
fig = plt.figure()
ax = fig.add_subplot(2,2,1)
ax = fig.add_subplot(2,2,2)
ax = fig.add_subplot(2,2,3)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(221)
ax = fig.add_subplot(222)
ax = fig.add_subplot(223)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
data = np.array([1,2,3])

plt.plot(data)
plt.show()
```

```python
# 두개의 그래프 순차적으로 그리기

fig = plt.figure()
x = [1,2,3]
y1 = [1,2,3]
y2 = [3,2,1]
ax1 = fig.add_subplot(211)
plt.plot(x, y1)

ax2 = fig.add_subplot(212)
plt.plot(x, y2)

plt.show()
```

```python

fig = plt.figure()
x = [1,2,3]
y1 = [1,2,3]
y2 = [3,2,1]
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.plot(x, y1)
ax2.plot(x, y2)

plt.show()
```

## plot 종류별 사용법


### bar plot

```python
# bar plot

fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0,1,2,3,4],[3,2,5,4,6])
plt.show()
```

```python
# bar plot

fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0,1,2,3,4],[3,2,5,4,6], width = 0.5, bottom = 2, align = 'edge')
plt.show()
```

### histogram

```python
# histogram
np.random.seed(1111)
fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500))
plt.show()
```

```python
# histogram
np.random.seed(1111)
fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins = 50)
plt.show()
```

```python
# histogram
np.random.seed(1111)
fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), [-2,-1,0,1,2,3])
plt.show()
```

```python
# histogram
np.random.seed(1111)
fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), np.linspace(-2,2,num = 40))
plt.show()
```

```python
# histogram
np.random.seed(1111)
fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), range = (-2,2), density = True)
plt.show()
```
